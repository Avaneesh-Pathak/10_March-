Q1: The Probability Mass Function (PMF) and Probability Density Function (PDF) are mathematical functions used to describe the probability distribution of a random variable.

The PMF is used for discrete random variables and gives the probability of each possible outcome. It assigns probabilities to individual values. For example, consider a fair six-sided die. The PMF of this random variable would give the probability of rolling each of the numbers 1, 2, 3, 4, 5, and 6.

The PDF, on the other hand, is used for continuous random variables. It represents the likelihood of the random variable taking on a particular value within a range. Unlike the PMF, which assigns probabilities to specific values, the PDF assigns probabilities to intervals. For example, if we have a continuous random variable representing the height of individuals in a population, the PDF would describe the probability of a person having a height within a given range, such as between 160 cm and 170 cm.

Q2: The Cumulative Density Function (CDF) is a function that gives the probability that a random variable takes on a value less than or equal to a given value. It provides a cumulative view of the probabilities.

For example, let's say we have a random variable representing the time it takes for a customer to complete a transaction at a store. The CDF would give the probability that a customer takes less than or equal to a certain amount of time. If we want to find the probability that a customer takes less than 5 minutes to complete the transaction, we can use the CDF to calculate that probability.

The CDF is used to determine probabilities for a range of values and can be helpful in understanding the overall distribution of a random variable.

2. Finite variance: The random variables have a finite variance, which measures the spread or variability of the variable. This assumption ensures that the sum or average of the variables remains well-behaved as the sample size increases.

3. Sample size: The Central Limit Theorem holds better as the sample size increases. While there is no fixed threshold, a general guideline is that the sample size should be reasonably large (often at least 30) for the theorem to provide a good approximation.

These assumptions, when met, allow the Central Limit Theorem to hold and enable the use of the normal distribution as an approximation for the sampling distribution of the mean. Violating these assumptions may lead to deviations from normality and affect the validity of the theorem's application.

Q1: Estimation statistics involves the process of estimating population parameters based on sample data. It aims to draw conclusions or make inferences about the population using information from a representative sample.

Point estimate: A point estimate is a single value that is used to estimate or approximate a population parameter. It is calculated from sample data and provides an estimate of the unknown population parameter. For example, using the sample mean as an estimate of the population mean is a point estimate.

Interval estimate: An interval estimate, also known as a confidence interval, provides a range of values within which the population parameter is likely to lie. It takes into account both the point estimate and the variability in the data. A confidence level is associated with the interval estimate, representing the probability that the interval contains the true population parameter. For example, an interval estimate could be "We are 95% confident that the population mean falls between X and Y."

```python
import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    # Calculate the standard error of the mean
    standard_error = sample_std_dev / math.sqrt(sample_size)
    
    # Calculate the margin of error (e.g., for a 95% confidence level, use a z-value of 1.96)
    margin_of_error = 1.96 * standard_error
    
    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound
```

Q8: The p-value is a probability value that measures the strength of evidence against the null hypothesis. It quantifies the likelihood of observing the sample data or more extreme data under the assumption that the null hypothesis is true.

In hypothesis testing, the p-value is compared to the chosen significance level (α). If the p-value is less than α, the results are considered statistically significant, and the null hypothesis is rejected. If the p-value is greater than or equal to α, there is insufficient evidence to reject the null hypothesis.

The significance of the p-value lies in its ability to help researchers make decisions and draw conclusions based on statistical evidence. A smaller p-value indicates stronger evidence against the null hypothesis and supports the alternative hypothesis. It allows researchers to determine the level of confidence they can have in their findings and helps in communicating the statistical significance of the results
